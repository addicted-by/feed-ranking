```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoded = encoder.fit_transform(df_clf['category'])
```

```python
df_clf['category'] = encoded
```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertForSequenceClassification, AdamW

class CustomDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data['text'].loc[idx]
        label = self.data['category'].loc[idx]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        input_ids = encoding['input_ids'].squeeze()
        attention_mask = encoding['attention_mask'].squeeze()
        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'label': label
        }

# Load and preprocess the DataFrame
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
max_length = 128  # Maximum sequence length for BERT input

# Create instances of the custom dataset for train and validation
dataset = CustomDataset(df_clf, tokenizer, max_length)
```

```python
dataset[0]
```

```python
from sklearn.model_selection import train_test_split


train_dataset, val_dataset = train_test_split(dataset, test_size=0.3)
```

```python
# Create data loaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
```

```python
# Load pre-trained BERT model and define the classifier
model_name = 'bert-base-uncased'  # Pre-trained BERT model
num_labels = 18  # Number of classification labels
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

# Set device (CPU or GPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define the optimizer and learning rate
optimizer = AdamW(model.parameters(), lr=2e-5)

# Training loop
num_epochs = 5
```

```python
num_labels
```

```python
from tqdm import tqdm
from IPython.display import clear_output
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

def train_epoch(model, tepoch):
    train_loss = 0.
    correct_predictions = 0
    for batch in tepoch:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        labels = batch['label'].to(device)

        optimizer.zero_grad()

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

        loss = outputs.loss
        train_loss += loss.item()


        logits = outputs.logits
        predicted_labels = torch.argmax(logits, dim=1)
        correct_predictions += (predicted_labels == labels).sum().item()
        loss.backward()
        optimizer.step()

    return train_loss / len(tepoch), correct_predictions / len(train_dataset)


def val_epoch(model, tepoch):
    val_loss = 0.
    correct_predictions = 0
    model.eval()

    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=labels
            )

            loss = outputs.loss
            val_loss += loss.item()

            logits = outputs.logits
            predicted_labels = torch.argmax(logits, dim=1)
            correct_predictions += (predicted_labels == labels).sum().item()

    return val_loss / len(tepoch), correct_predictions / len(val_dataset)

def train(model, num_epochs):
    losses, val_losses, accuracies, val_accuracies = [], [], [], []
    for epoch in tqdm(range(num_epochs)):
        with tqdm(train_loader, unit="batch") as tepoch:
            tepoch.set_description(f"Epoch {epoch + 1}")
            model.train()
            loss, accuracy = train_epoch(model, tepoch)
            losses.append(loss)
            accuracies.append(accuracy)
            

        # Validate the model
        with tqdm(val_loader, unit='batch') as tvalepoch:
            val_loss, val_accuracy = val_epoch(model, tvalepoch)
            val_losses.append(val_loss)
            val_accuracies.append(val_accuracy)

        # show plots
        clear_output()
        fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15,10))
        fig.suptitle(f"Epoch {epoch+1}")
        axes[0].plot(losses, label='Train loss')
        axes[0].plot(val_losses, label='Validation loss')
        axes[0].legend()
        axes[0].set_title("Losses during training")

        axes[1].plot(losses, label='Train loss')
        axes[1].plot(val_losses, label='Validation loss')
        axes[1].legend()
        axes[1].set_title("Accuracy during training")

        plt.show()


    history = losses, val_losses, accuracies, val_accuracies

    return history
```

```python
train(model, num_epochs)
```
